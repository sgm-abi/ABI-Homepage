```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def find_broken_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('a')
    print(f"Links insgesamt: {len(links)}")
    
    broken_links = []
    for link in links:
        href = link.get('href')
        print(href)
        if href.startswith('http'):
            print("check")
            link_response = requests.head(href)
            if link_response.status_code >= 400:
                broken_links.append(href)
    
    return broken_links
```

```python
url = 'https://sgm-abi.de'  # Ersetze dies mit der gewünschten URL
broken_links = find_broken_links(url)

print(f"Defekte Links: {len(broken_links)}")
for link in broken_links:
    print(link)
```
